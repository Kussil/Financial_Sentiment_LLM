```python
import pandas as pd
import os

# List of CSV files to be converted in the current working directory
csv_files = [
    'ProQuest_Articles.csv',
    'SEC_Filings.csv',
    'Investment_Research_Part2.csv',
    'Earnings_QA.csv',
    'Investment_Research_Part1.csv',
    'Earnings_Presentations.csv'
]

def convert_csv_to_parquet(file_path):
    try:
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Define the Parquet file path
        parquet_file_path = file_path.replace('.csv', '.parquet')
        
        # Save the DataFrame as a Parquet file
        df.to_parquet(parquet_file_path, index=False)
        print(f"Converted {file_path} to {parquet_file_path}")
    except Exception as e:
        print(f"Failed to convert {file_path}: {e}")

# Convert each CSV file to Parquet
for csv_file in csv_files:
    convert_csv_to_parquet(csv_file)
```
